# Problem 1:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 2 * np.pi, 40)

# True sine function
y_true = np.sin(x)

# Noise Generation
noise = np.random.normal(0, 0.05, size=x.shape)
y_noisy = y_true + noise

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, y_true, label='Sine Function', color='blue')
plt.scatter(x, y_noisy, label='Noisy Data', color='red', s=10)
plt.title('Sine Function vs Noisy Data')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

# Problem 2:

```python
def is_monotonic(arr):
    increasing = True
    decreasing = True

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            increasing = False
        if arr[i] > arr[i - 1]:
            decreasing = False

    if increasing:
        return "Monotonic Increasing"
    elif decreasing:
        return "Monotonic Decreasing"
    else:
        return "Not Monotonic"

# Test examples
arr1 = [1,2,3,3,3,3,3,4,5,6,7]
arr2 = [7,6,5,4,3,3,3,3,3,2,1]
arr3 = [1,5,4,3,3,3,3,3,2,6,7]

print(f"{arr1} --> {is_monotonic(arr1)}")
print(f"{arr2} --> {is_monotonic(arr2)}")
print(f"{arr3} --> {is_monotonic(arr3)}")
```
