# Part one: Analytical Method

## Slicing the training dataset

```python
x_train=Ds_train[:,0]
y_train=Ds_train[:,1]
```

```python
# Calculating the necessary parameters
n = len(x_train)
xi_train = np.sum(x_train)
yi_train = np.sum(y_train)
xi2_train = np.sum(x_train**2)
xiyi_train = np.sum(x_train*y_train)

# Calculating the w values
w1 = (n*xiyi_train-xi_train*yi_train)/(n*xi2_train-xi_train**2)
w0 = (yi_train-w1*xi_train)/n
w=np.array([w0,w1])
print(w)

# Predicting the y values
X_train = np.c_[np.ones(n),x_train]
y_pred_train = X_train.dot(w)
```

### Defining the RMSE function

```python
def RMSE(Y_pred,Y):
    return np.sqrt(np.mean((Y_pred-Y)**2))
```

### Plotting the results for each dataset

```python
plt.scatter(x_train,y_train,color='red',label="training data")
plt.plot(x_train,y_pred_train,'-',label="fitted line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training Dataset')
plt.legend()
plt.show()
print(RMSE(y_pred_train,y_train))
```

```python
x_valid = Ds_valid[:,0]
y_valid = Ds_valid[:,1]
plt.scatter(x_valid,y_valid,color='orange',label="Validation data")
plt.plot(x_train,y_pred_train,'-',label="fitted line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Validation Dataset')
plt.legend()
plt.show()
print(RMSE(y_pred_train,y_valid))
```

```python
x_test = Ds_test[:,0]
y_test = Ds_test[:,1]
plt.scatter(x_test,y_test,color='black',label="Testing data")
plt.plot(x_train,y_pred_train,'-',label="fitted line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Testing Dataset')
plt.legend()
plt.show()
print(RMSE(y_pred_train,y_test))
```

# Part two: Polynomial Method

## Finding the optimum degree

```python
# The maximum degree
D=10

# Setting up the X arrays
X_poly_train = np.c_[np.ones(n)]
X_poly_valid = np.c_[np.ones(n)]
rmse_train=[]
rmse_valid=[]
for i in range(1,D+1):
    X_poly_train = np.c_[X_poly_train,x_train**i]
    # Calculating the w values for the training dataset
    w = la.inv(X_poly_train.T.dot(X_poly_train)).dot(X_poly_train.T).dot(y_train)
    # Predicting the y values for the training dataset
    y_pred_train = X_poly_train.dot(w)
    # Calculating the RMSE for the training dataset
    rmse_train.append(RMSE(y_pred_train,y_train))

    # Repeating the proccess for the validation dataset
    X_poly_valid = np.c_[X_poly_valid,x_valid**i]
    y_pred_valid = X_poly_valid.dot(w)
    rmse_valid.append(RMSE(y_pred_valid,y_valid))

    # Storing the degree at which validation RMSE is minimum
    if i == 1:
      min_rmse = RMSE(y_pred_valid,y_valid)
      min_deg = i
    elif min_rmse > RMSE(y_pred_valid,y_valid):
      min_rmse = RMSE(y_pred_valid,y_valid)
      min_deg = i


# Printing the minimum validation degree
print(f"The optimum degree is {min_deg}")
```

### Plotting the result

```python
plt.plot(range(1,D+1),rmse_train,label='Train')
plt.plot(range(1,D+1),rmse_valid,color='black',label='Valid')
plt.xlabel('Degree')
plt.ylabel('RMSE')
plt.legend()
plt.title('RMSE vs Degree')
plt.show()
```

## Using the Optimum Degree to:

## Fitting the model using the training data

```python
# Setting up the x array
x_fit_train = x_train
x_fit_poly_train = np.c_[np.ones(len(x_fit_train))]
for i in range(1,min_deg+1):
  x_fit_poly_train = np.c_[x_fit_poly_train,x_fit_train**i]

# Calculating w
w = la.inv(x_fit_poly_train.T.dot(x_fit_poly_train)).dot(x_fit_poly_train.T).dot(y_train)

# Calculting the predicted y values
y_fit_train = x_fit_poly_train.dot(w)
print(w)
```

### Calculating the RMSE and plotting the result for the training dataset

```python
# Setting up the X array
x_plot = np.linspace(x_train.min(), x_train.max(), 100)
X_plot_poly = np.c_[np.ones(len(x_plot))]
for i in range(1, min_deg + 1):
    X_plot_poly = np.c_[X_plot_poly, x_plot**i]

# Calculating the y values
y_plot = X_plot_poly.dot(w)

# Plotting the result
plt.scatter(x_train, y_train, color='red', label='Training data')
plt.plot(x_plot, y_plot, '-', label=f'Fitted polynomial model \n using the training dataset (Degree {min_deg})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training Dataset')
plt.legend()
plt.show()
print(f"RMSE = {RMSE(y_fit_train, y_train)}")
```

### Graphing the result for the Validation dataset and calculating the RMSE

```python
# Plotting the result
plt.scatter(x_valid,y_valid,color='orange',label='Validation data')
plt.plot(x_plot,y_plot,'-',label=f'Fitted polynomial model \n using the training dataset (Degree {min_deg})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Validation Dataset')
plt.legend()
plt.show()

X_valid_poly = np.c_[np.ones(len(x_valid))]
for i in range(1, min_deg + 1):
    X_valid_poly = np.c_[X_valid_poly, x_valid**i]
y_pred_valid = X_valid_poly.dot(w)

# Calculating the RMSE
print(f"RMSE = {RMSE(y_pred_valid,y_valid)}")
```

### Calculating the RMSE and graphing the result for the Testing dataset

```python
# Plotting the result
plt.scatter(x_test,y_test,color='black',label='Testing data')
plt.plot(x_plot,y_plot,'-',label=f'Fitted polynomial model \n using the training dataset (Degree {min_deg})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Testing Dataset')
plt.legend()
plt.show()

X_test_poly = np.c_[np.ones(len(x_test))]
for i in range(1, min_deg + 1):
    X_test_poly = np.c_[X_test_poly, x_test**i]
y_pred_test = X_test_poly.dot(w)

# Calculating the RMSE
print(f"RMSE = {RMSE(y_pred_test,y_test)}")
```
